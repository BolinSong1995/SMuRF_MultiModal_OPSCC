```python
import pickle
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Load the pickle file
pickle_path = r"C:\Users\bsong47\OneDrive - Emory University\Documents\code\raptomics\checkpoints\raw_images\fused_attention_multitask_100_0.002_raptomic\fused_attention_multitask_100_0.002_raptomic_pred_test.pkl"
with open(pickle_path, 'rb') as file:
    data = pickle.load(file)

# Extract prediction scores and ground truth
pred_scores = data[0]  # First element: prediction scores
ground_truth = data[2]  # Third element: binary ground truth

# If they are tensors, move them to the CPU, detach from graph, and convert to NumPy arrays
if hasattr(pred_scores, 'cpu'):
    pred_scores = pred_scores.detach().cpu().numpy()

if hasattr(ground_truth, 'cpu'):
    ground_truth = ground_truth.detach().cpu().numpy()

# Compute ROC curve and AUC
fpr, tpr, thresholds = roc_curve(ground_truth, pred_scores)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve with adjusted size and 300 dpi resolution
plt.figure(figsize=(6, 6), dpi=300)
plt.plot(fpr, tpr, color='darkorange', lw=4, label=f'AUC = {roc_auc:.2f}')  # Bold line (lw=3)
plt.plot([0, 1], [0, 1], color='navy', lw=3, linestyle='--')  # Reference line

# Increase font sizes for axes and legend, set Arial font
plt.xlabel('False Positive Rate', fontname='Arial', fontsize=16)  # Larger font for x-axis
plt.ylabel('True Positive Rate', fontname='Arial', fontsize=16)  # Larger font for y-axis
plt.xticks(fontname='Arial', fontsize=14)
plt.yticks(fontname='Arial', fontsize=14)

# Increase legend font size and set Arial font explicitly
legend = plt.legend(loc="lower right", prop={'family': 'Arial', 'size': 21})

# Save the figure to the specified path with 300 dpi
save_path = r"C:\Users\bsong47\OneDrive - Emory University\Documents\Raptomic_OPSCC_paper\CancerDiscovery\figures\roc_curve_test.png"
plt.savefig(save_path, dpi=300)

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import shap
from lifelines import CoxPHFitter
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Load the data
file_path = r"C:\Users\bsong47\OneDrive - Emory University\Documents\Raptomic_OPSCC_paper\CancerDiscovery\data_table_combined_test.csv"
data = pd.read_csv(file_path)

# Prepare the data
# Encode categorical variables
categorical_columns = ['T-stage', 'N-stage', 'Sex', 'Treatment']
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le

# Define the feature set and target variable
X = data[['Age', 'Smoking PY', 'SMuRF'] + categorical_columns]
y = data[['DFS', 'DFS_censor']]
combined_data = pd.concat([X, y], axis=1)
# Fit the Cox proportional hazards model
cph = CoxPHFitter()
cph.fit(combined_data, duration_col='DFS', event_col='DFS_censor')
print(combined_data)

# Display the summary of the Cox model
print(cph.summary)

# Calculate SHAP values
explainer = shap.Explainer(cph.predict_partial_hazard, X)
shap_values = explainer(X)

# Set the font to Arial and make it bold
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.weight'] = 'bold'  # Set font weight to bold

# Plot the SHAP values using a beeswarm plot
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, X, plot_type="dot", show=False)  # Prevents auto display

# Set the title with bold font
# plt.title('Beeswarm Plot of SHAP Values in Multivariate Cox Analysis (DFS)', fontsize=16, fontweight='bold')

# Save the plot
output_path = r'C:\Users\bsong47\OneDrive - Emory University\Documents\Raptomic_OPSCC_paper\CancerDiscovery\figures\beeswarm.png'
plt.savefig(output_path, dpi=300, bbox_inches='tight')

plt.close()  # Close the figure after saving

```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.cm as cm

# Calculate mean absolute SHAP values for each feature
mean_shap_values = np.abs(shap_values.values).mean(axis=0)

# Create a DataFrame to hold feature names and their mean SHAP values
features = X.columns
mean_shap_df = pd.DataFrame({'Feature': features, 'Mean SHAP Value': mean_shap_values})

# Sort by Mean SHAP Value
mean_shap_df.sort_values(by='Mean SHAP Value', ascending=False, inplace=True)

# Normalize mean SHAP values for color mapping with the 'Blues' colormap
norm = plt.Normalize(mean_shap_df['Mean SHAP Value'].min(), mean_shap_df['Mean SHAP Value'].max())
colors = cm.Blues(norm(mean_shap_df['Mean SHAP Value']))  # Use the 'Blues' colormap

# Set the font to Arial and bold
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.weight'] = 'bold'

# Custom autopct function to change the color of the proportion value for SMuRF
def custom_autopct(pct, labels):
    # Match label to "SMuRF" and set it to white
    label = labels[custom_autopct.counter]
    custom_autopct.counter += 1
    if label == 'smurf':
        return f'{pct:.1f}%' if pct > 0 else ''
    else:
        return f'{pct:.1f}%' if pct > 0 else ''

# Initialize a counter for custom_autopct
custom_autopct.counter = 0

# Plot a pie chart
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(mean_shap_df['Mean SHAP Value'], labels=mean_shap_df['Feature'], 
                                   autopct=lambda p: custom_autopct(p, mean_shap_df['Feature']),
                                   startangle=140, colors=colors, textprops={'fontsize': 20})

# Set the SMuRF variable's percentage text to white
for i, autotext in enumerate(autotexts):
    if mean_shap_df['Feature'].iloc[i] == 'SMuRF':
        autotext.set_color('white')  # Set SMuRF proportion to white
    else:
        autotext.set_color('black')  # All other proportions remain black

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Save the pie chart
output_path = r'C:\Users\bsong47\OneDrive - Emory University\Documents\Raptomic_OPSCC_paper\CancerDiscovery\figures\shap_mean_pie_chart.png'
plt.savefig(output_path, dpi=300, bbox_inches='tight')

plt.show()  # Display the pie chart
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import shap
import matplotlib.pyplot as plt

# Load the data
file_path = r"C:\Users\bsong47\OneDrive - Emory University\Documents\Raptomic_OPSCC_paper\CancerDiscovery\data_table_combined_test.csv"
data = pd.read_csv(file_path)

# Prepare the data
# Encode categorical variables
categorical_columns = ['T-stage', 'N-stage', 'Sex', 'Treatment']
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le

# Define the feature set and target variable
X = data[['Age', 'Smoking PY', 'SMuRF'] + categorical_columns]
y = data['grade']  # 'grade' is the target variable (endpoint)

# Initialize and fit a logistic regression model
model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X, y)

# Predict the target and display a classification report
y_pred = model.predict(X)
print(classification_report(y, y_pred))

# Display the confusion matrix
print(confusion_matrix(y, y_pred))

# Calculate SHAP values for the logistic regression model
explainer = shap.Explainer(model, X)
shap_values = explainer(X)

# Set the font to Arial and bold
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.weight'] = 'bold'

# Plot the SHAP values using a beeswarm plot
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, X, plot_type="dot", show=False)

# Save the plot
# output_path = r'C:\Users\bsong47\OneDrive - Emory University\Documents\Raptomic_OPSCC_paper\CancerDiscovery\figures\logistic_shap_beeswarm.png'
plt.savefig(output_path, dpi=300, bbox_inches='tight')

plt.close()  # Close the figure after saving
```
